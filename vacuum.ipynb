```python
import random

row, col = random.randint(3, 6), random.randint(3, 6)

mat = []
for i in range(row):
    mat.append([random.randint(0, 2) for j in range(col)])

def possible_move(x, y, mat):
    move = []
    if x > 0 and mat[x-1][y] != 2:
        move.append((-1, 0)) # Lên
    if x < row-1 and mat[x+1][y] != 2:
        move.append((1, 0))  # Xuống
    if y > 0 and mat[x][y-1] != 2:
        move.append((0, -1)) # Trái
    if y < col-1 and mat[x][y+1] != 2:
        move.append((0, 1))  # Phải
    return move

def check_floor(mat):
    for i in range(row):
        for j in range(col):
            if mat[i][j] == 1:
                return False
    return True

def main():
    count = 0
    visited = set()

    while True:
        x, y = random.randint(0, row-1), random.randint(0, col-1)
        if mat[x][y] != 2:
            vacuum_pos = [x, y]
            break

    print("Sàn ban đầu:")
    for i in mat:
        print(i)

    print(f"\nBắt đầu tại: {vacuum_pos[0]}, {vacuum_pos[1]}")

    while (not check_floor(mat)) and count < 100:
        current_state = (tuple(vacuum_pos), tuple(tuple(r) for r in mat))

        if current_state in visited:
            print("Lặp lại và dừng")
            break

        visited.add(current_state)
        x, y = vacuum_pos[0], vacuum_pos[1]

        if mat[x][y] == 1:
            print(f"Hút bụi tại {x}, {y}")
            mat[x][y] = 0
        else:
            moves = possible_move(x, y, mat)

            if not moves:
                print("Kẹt vật cản")
                break

            direction = random.choice(moves)
            vacuum_pos[0] += direction[0]
            vacuum_pos[1] += direction[1]
            print(f"Di chuyển đến: {vacuum_pos[0]}, {vacuum_pos[1]}")

        count += 1

    if check_floor(mat):
        print("Sàn đã sạch")
    elif count >= 100:
        print("Dừng vì đã chạy quá 100 bước")

if __name__ == "__main__":
    main()
```
