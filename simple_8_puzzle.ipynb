```python
import random

GOAL_STATE = (1, 2, 3, 5, 6, 4, 8, 7, 0)

def get_possible_states(state):
    next_states = []
    state_list = list(state)
    empty_idx = state_list.index(0)
    row, col = divmod(empty_idx, 3)

    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    for dr, dc in directions:
        r, c = row + dr, col + dc
        if 0 <= r < 3 and 0 <= c < 3:
            new_state = state_list[:]
            target_idx = r * 3 + c
            new_state[empty_idx], new_state[target_idx] = new_state[target_idx], new_state[empty_idx]
            next_states.append(tuple(new_state))

    return next_states

def solve_with_no_repeat(initial_state):
    current_state = tuple(initial_state)
    visited = {current_state}
    path = [current_state]

    print(f"Bắt đầu tại: {current_state}")

    step = 0
    while current_state != GOAL_STATE:
        step += 1
        possible_next_states = get_possible_states(current_state)

        next_state = random.choice(possible_next_states)

        print(f"Bước {step}: Di chuyển tới {next_state}")

        if next_state in visited:
            print(f"==> DỪNG LẠI: Trạng thái {next_state} đã từng xuất hiện trước đó!")
            return False, path

        visited.add(next_state)
        path.append(next_state)
        current_state = next_state

    print("==> THÀNH CÔNG: Đã tìm thấy trạng thái đích!")
    return True, path

def is_solvable(state):
    flat = [n for n in state if n != 0]
    inversions = sum(flat[i] > flat[j] for i in range(len(flat)) for j in range(i + 1, len(flat)))
    return inversions % 2 == 0

nodes = list(range(9))
while True:
    random.shuffle(nodes)
    start_node = tuple(nodes)
    if is_solvable(start_node): # Chỉ lấy mảng nếu toán học cho phép giải được
        break

success, final_path = solve_with_no_repeat(start_node)

print(f"\nTổng số bước đã đi: {len(final_path) - 1}")
```
