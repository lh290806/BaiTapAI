# PEAS

P: Khả năng tối ưu nhất để sàn nhà sạch

E: Ma trận mxn bthi các số 0 và 1 (0 là sạch, 1 là sàn bẩn)

A: Các hàm để agent hiểu và hành động

S: Cơ chế, hàm để agent đọc dữ liệu (vi trí, trạng thái)

```python
import random

row , col = random.randint(2,5), random.randint(2,5)

mat = []
for i in range(row):
    mat.append([random.randint(0, 1) for j in range(col)])

def possible_move(x, y):
    move = []
    if x > 0: move.append((-1, 0))
    if x < row-1: move.append((1, 0))
    if y > 0: move.append((0, -1))
    if y < col-1: move.append((0, 1))
    return move

def check_floor(mat):
    for i in range (0, row):
        for j in range (0, col):
            if mat[i][j] == 1:
                return False
    return True

def main():
    count = 0
    visited = set()
    vacuum_pos = [random.randint(0,row-1), random.randint(0,col-1)]

    for r in mat:
        print(r)

    print(f"\nBắt đầu: {vacuum_pos[0]}, {vacuum_pos[1]}")

    while (not check_floor(mat)) and count < 100:
        current_state = (tuple(vacuum_pos), tuple(tuple(r) for r in mat))

        if current_state in visited:
            print("Lặp lại và dừng!")
            break

        visited.add(current_state)

        x, y = vacuum_pos[0], vacuum_pos[1]

        if mat[x][y] == 1:
            print(f"Hút bụi tại {vacuum_pos[0]}, {vacuum_pos[1]}")
            mat[x][y] = 0
        else:
            moves = possible_move(x, y)
            direction = random.choice(moves)
            vacuum_pos[0] += direction[0]
            vacuum_pos[1] += direction[1]
            print(f"Di chuyển đến: {vacuum_pos[0]}, {vacuum_pos[1]}")
        count += 1

    if check_floor(mat):
        print("Sàn đã sạch hoàn toàn!")
    elif count >= 100:
        print("Dừng vì đã chạy quá 100 bước.")

if __name__ == "__main__":
    main()
```
