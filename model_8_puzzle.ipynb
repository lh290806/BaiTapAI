#Tập luật
Luật 1: Kiểm tra đã giải đc chưa

Luật 2: Lưu lại các trạng thái đã có

Luật 3: Dự đoán đường đi

    Nếu đã có thì cấm đi đường đó

    Nếu ko có thì thêm vào danh sách có thể đi

Luật 4: Hàng động

    Nếu danh sách trống thì dừng vì ko còn đường đi

    Nếu danh sách có đường đi thì random chọn 1

```python
import random

GOAL = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

def possible_move(x, y):
    moves = []
    if x > 0:
        moves.append((-1, 0)) # Lên
    if x < 2:
        moves.append((1, 0))  # Xuống
    if y > 0:
        moves.append((0, -1)) # Trái
    if y < 2:
        moves.append((0, 1))  # Phải
    return moves

def check_goal(mat):
    return mat == GOAL

def find_zero(mat):
    for i in range(3):
        for j in range(3):
            if mat[i][j] == 0:
                return i, j

def simulate_move(mat, x, y, dx, dy):
    new_mat = [list(row) for row in mat]
    new_x, new_y = x + dx, y + dy
    new_mat[x][y], new_mat[new_x][new_y] = new_mat[new_x][new_y], new_mat[x][y]
    return new_mat

def main():
    mat = [
        [1, 2, 3],
        [4, 5, 0],
        [7, 8, 6]
    ]

    count = 0
    visited = set()

    print("Bảng ban đầu:")
    for i in mat:
        print(i)

    while (not check_goal(mat)):
        current_state = tuple(tuple(r) for r in mat)
        visited.add(current_state)

        x, y = find_zero(mat)
        moves = possible_move(x, y)

        unvisited_moves = []

        for dx, dy in moves:
            simulated_mat = simulate_move(mat, x, y, dx, dy)
            sim_tuple = tuple(tuple(r) for r in simulated_mat)

            if sim_tuple not in visited:
                unvisited_moves.append((dx, dy, simulated_mat))

        if not unvisited_moves:
            print("\nKẹt")
            break

        chosen_dx, chosen_dy, next_mat = random.choice(unvisited_moves)

        mat = next_mat
        count += 1

        print(f"\nBước {count}: Di chuyển ô trống đến {x + chosen_dx}, {y + chosen_dy}")
        for i in mat: print(i)

    if check_goal(mat):
        print("\nHoàn thành!")
    else:
        print(f"\nTổng bước: {count}")

if __name__ == "__main__":
    main()
```
